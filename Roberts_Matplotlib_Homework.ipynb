```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

all_data = pd.merge(city_df, ride_df, on="city", how="right")

# Display the data table for preview
all_data
```

```python
#Insert 3 tables. Then use this data x 3 types - suburban and rural. 

urban_cities = all_data[all_data["type"] == "Urban"]
urban_cities
```

## Bubble Plot of Ride Sharing Data

```python
all_data.dtypes
```

```python
all_data["type"].unique()
```

```python
#Get total number of rides per city - Shows data range from 0 to 39

rides_percity = all_data["city"].value_counts()
rides_percity.head()
```

```python
#Get average fare per city

grouped_city_df = all_data.groupby("city")
grouped_city_df.head()
```

```python
rides_per_city = grouped_city_df["ride_id"].count()
rides_per_city
```

```python
avg_fare_city = grouped_city_df["fare"].mean()
avg_fare_city.head()
```

```python
#Get range for average fare by city
avg_fare_city.min()
```

```python
#Get range for average fare by city
avg_fare_city.max()
```

```python
#Get total rides by city

total_rides_city = grouped_city_df["ride_id"].count()
total_rides_city.head()
```

```python
#Get range for total rides by city  - maximum value

total_rides_city.max()
```

```python
#Get range for total rides  by city - minimum value

total_rides_city.min()
```

```python
#Get type by city

type_city = grouped_city_df["type"].unique()
type_city
```

```python
# Drivers per city via pivot

driver_city_unique = pd.pivot_table(all_data, index="city", values= "driver_count").sort_values("driver_count")
driver_city_unique
```

```python
#New data frame with all the variables / by city

all_bubble_data = ({    "Average Fare per City": avg_fare_city,
                         "Total Rides per City": rides_per_city,
                        "Type": type_city
                         })
final_numbers_df = pd.DataFrame(all_bubble_data)
final_numbers_df
```

```python
pivot_plus_df = pd.merge(final_numbers_df, driver_city_unique, on="city", how="right")

# Display the data table for preview
pivot_plus_df
```

```python
# Obtain the x and y coordinates for each of the three city types

x_value = pivot_plus_df["Total Rides per City"]
y_axis = pivot_plus_df["Average Fare per City"]
x_limit = 40
plt.ylim(20, 45)
plt.xlim(1, x_limit)

x_axis = np.arange(0, x_limit, 10)

# Build the scatter plots for each 3 city types
# 1. Average fare/city range from 20 to 45 and is Y axis
# 2. Driver counts range 1 to 75
# 3. Total number of rides per city range is 0 to 40 and is X axis

driver = pivot_plus_df["driver_count"]

plt.scatter(x_value, y_axis, marker="o", facecolors="blue", edgecolors="black", s=driver*3, alpha=0.75)


#Circle size = driver count per city


# Incorporate the other graph properties

plt.title("Pyber Rides by Region 2016")
plt.xlabel("Total Rides per City")
plt.ylabel("Average Fare")

# Create a legend

# Adds a legend and sets its location to the lower right

plt.legend(loc="upper right")

# Incorporate a text label regarding circle size
```

```python
#Get just urban graph point
type_urban = pivot_plus_df.loc[pivot_plus_df["Type"] == "Urban"]
type_urban.head()
```

```python
type_rural = pivot_plus_df.loc[pivot_plus_df["Type"] == "Rural"]
type_rural.head()
```

```python
type_suburban = pivot_plus_df.loc[pivot_plus_df["Type"] == "Suburban"]
type_suburban.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
#Type = Suburban
x_suburban = type_suburban["Total Rides per City"]
y_suburban = type_suburban["Average Fare per City"]
x_limit = 40
plt.ylim(20, 45)
plt.xlim(1, x_limit)
x_axis = np.arange(0, x_limit, 10)

#Type Urban
x_urban = type_urban["Total Rides per City"]
y_urban = type_urban["Average Fare per City"]
x_limit = 40
plt.ylim(20, 45)
plt.xlim(1, x_limit)
x_axis = np.arange(0, x_limit, 10)

#Type Rural
x_rural = type_rural["Total Rides per City"]
y_rural = type_rural["Average Fare per City"]
x_limit = 40
plt.ylim(20, 45)
plt.xlim(1, x_limit)
x_axis = np.arange(0, x_limit, 10)

# Build the scatter plots for each 3 city types
# 1. Average fare/city range from 20 to 45 and is Y axis
# 2. Driver counts range 1 to 75
# 3. Total number of rides per city range is 0 to 40 and is X axis

driver = pivot_plus_df["driver_count"]

plt.scatter(x_suburban, y_suburban, marker="o", facecolors="blue", edgecolors="black", s=driver*5, alpha=0.5, label = "Suburban")
plt.scatter(x_urban, y_urban, marker="o", facecolors="green", edgecolors="black", s=driver*5, alpha=0.5, label = "Urban")
plt.scatter(x_rural, y_rural, marker="o", facecolors="red", edgecolors="black", s=driver*5, alpha=0.5, label = "Rural")

#Circle size = driver count per city


# Incorporate the other graph properties

plt.title("Pyber Rides by Region 2016")
plt.xlabel("Total Rides per City")
plt.ylabel("Average Fare")

plt.grid(True)

# Create a legend
# Adds a legend and sets its location to the lower right

plt.legend(loc="upper right", fontsize = "small", title = "City Types", labelspacing = 0.5) 

# Incorporate a text label regarding circle size
```

## Total Fares by City Type

```python
# Calculate Type Percents
#Number of total fares per city (sum) divided by number of cities of same type

Fares_percent_city_Suburban = type_suburban["Average Fare per City"].sum()
Fares_percent_city_Suburban
```

```python
Fares_percent_city_Urban = type_urban["Average Fare per City"].sum()
Fares_percent_city_Urban
```

```python
Fares_percent_city_rural = type_rural["Average Fare per City"].sum()
Fares_percent_city_rural
```

```python
type_percent = 100 * all_data.groupby(["type"]).sum()["fare"] / all_data["fare"].sum()
type_percent
```

```python
# Build Pie Chart

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]
plt.title("% Total Fares by City Type")

# The values of each section of the pie chart
# sizes = [Fares_percent_city_Suburban['Average Fare per City'], Fares_percent_city_Urban['Average Fare per City'],
#          Fares_percent_city_rural['Average Fare per City']]

# The colors of each section of the pie chart
colors = ["yellow", "lightblue", "pink"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.pie(type_percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)

# Save Figure

plt.savefig("Pie_Faretype.png")
```
